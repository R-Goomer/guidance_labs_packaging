```python
# create a blank html file with date and random word in the name
import datetime
import random
import string

def create_blank_html_file():
    # Get the current date
    current_date = datetime.datetime.now().strftime("%Y-%m-%d")
    
    # Generate a random word
    random_word = ''.join(random.choices(string.ascii_lowercase, k=5))
    
    # Create the filename
    filename = f"/Users/rushilgoomer/Desktop/guidance_labs_blogs/posts/{current_date}_{random_word}.html"
    
    # Create a blank html file
    with open(filename, 'w') as file:
        file.write("<!DOCTYPE html>\n<html>\n<head>\n<title>Blank HTML</title>\n</head>\n<body>\n</body>\n</html>")
    
    print(f"Created file: {filename}")

create_blank_html_file()
```
